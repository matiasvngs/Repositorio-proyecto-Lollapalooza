# Aquí importé la libreria pandas para trabajar con mi base de datos y poder combinar variables para formar una tabla de datos para mi gráfico y subí mi base de datos en cvs en la carpeta y pedí que leea mi archivo cvs. Además, le pedi que limpiara los espacios invisibles en los nombres de columnas para que no salgan errores y después y pusé df.head () para verificar que se ejecutara bien.

```python
import pandas as pd

df = pd.read_csv("/content/Ediciones.csv", sep=";")
df.columns = df.columns.str.strip()
df["año"] = df["año"].astype(str)
df.head()
```

# Aquí transforme el formato a largo porque la base original estaba en ancho y esto lo logré a través de df.melt , también le dije las columnas que quería combinar y con replace cambié los nombres para que en el gráfico se vea más ordenado. También le pedi que el eje x que era año sea texto. y le pedi con df_long.head() que me visulizara.

```python
df_long = df.melt(
    id_vars=["año"],
    value_vars=[
        "n_artistas_mujeres",
        "n_artistas_hombres",
        "n_bandas_mujeres",
        "n_bandas_hombres",
        "n_bandas_mixtas"
    ],
    var_name="categoria",
    value_name="cantidad"
)

df_long["categoria"] = df_long["categoria"].replace({
    "n_artistas_mujeres": "Artistas Mujeres",
    "n_artistas_hombres": "Artistas Hombres",
    "n_bandas_mujeres": "Bandas Mujeres",
    "n_bandas_hombres": "Bandas Hombres",
    "n_bandas_mixtas": "Bandas Mixtas"
})
df_long.head()
```

# Aquí empecé a crear el gráfico e hice muchas versiones que fui borrando, para llegar a esta de abajo. Pusé alt.chart para crear el gráfico, .mark_bar para que fuera de barras y .encode para que las columnas estén conectados con los ejes y le dije que quería que se viera en el eje y e x y los titulé. Luego en el color, quería escoger mis propios colores, así que busqué sus códigos para atribuirselo a cada categoría y agregué la función de tooltip para poder tener interración pasando el mouse. Por último, agregué las propiedas del gráfico y lo ejecuté.

```python
import altair as alt

grafico = alt.Chart(df_long).mark_bar().encode(
    x=alt.X("año:N", title="Año"),
    y=alt.Y("cantidad:Q", title="Cantidad de participantes"),
    color=alt.Color(
        "categoria:N",
        title="Categoría",
        scale=alt.Scale(
            domain=[
                "Artistas Hombres",
                "Artistas Mujeres",
                "Bandas Hombres",
                "Bandas Mujeres",
                "Bandas Mixtas"
            ],
            range=[
                "#58b4cb",
                "#e85343",
                "#43ace8",
                "#e87a43",
                "#c9d849"
            ]
        )
    ),
    tooltip=["año", "categoria", "cantidad"]
).properties(
    width=700,
    height=400,
    title="Participación por género y tipo en Lollapalooza Chile (2011–2025)"
)

grafico
```
